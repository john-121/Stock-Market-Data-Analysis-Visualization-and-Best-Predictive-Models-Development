```python
#Import Necessary Libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
# Loading cleaned stock data
nike_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_nike_data.csv')
adidas_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_adidas_data.csv')
pepsico_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_pepsi_data.csv')
sony_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_sony_data.csv')
```

```python
# Ensure the 'Date' column is correctly formatted
nike_data['Date'] = pd.to_datetime(nike_data['Date'])
adidas_data['Date'] = pd.to_datetime(adidas_data['Date'])
pepsico_data['Date'] = pd.to_datetime(pepsico_data['Date'])
sony_data['Date'] = pd.to_datetime(sony_data['Date'])

# Define start dates for different periods
last_year_date = pd.Timestamp.now() - pd.DateOffset(years=1)
last_month_date = pd.Timestamp.now() - pd.DateOffset(months=1)
last_week_date = pd.Timestamp.now() - pd.DateOffset(weeks=1)
last_five_days_date = pd.Timestamp.now() - pd.DateOffset(days=5)

# Calculate summary statistics for different periods
def summary_statistics(df, start_date):
    data_period = df[df['Date'] >= start_date]
    return data_period.describe()

# Summary statistics for the last year
nike_summary_year = summary_statistics(nike_data, last_year_date)
adidas_summary_year = summary_statistics(adidas_data, last_year_date)
pepsico_summary_year = summary_statistics(pepsico_data, last_year_date)
sony_summary_year = summary_statistics(sony_data, last_year_date)

print("Nike - Last Year Summary:")
print(nike_summary_year)
print("Adidas - Last Year Summary:")
print(adidas_summary_year)
print("PepsiCo - Last Year Summary:")
print(pepsico_summary_year)
print("Sony - Last Year Summary:")
print(sony_summary_year)
```

```python
# Function to plot price trends
def plot_price_trends(df, start_date, title):
    plt.figure(figsize=(12, 6))
    plt.plot(df[df['Date'] >= start_date]['Date'], df[df['Date'] >= start_date]['Price'], label='Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

# Plot price trends for the last year
plot_price_trends(nike_data, last_year_date, 'Nike - Price Trend (Last Year)')
plot_price_trends(adidas_data, last_year_date, 'Adidas - Price Trend (Last Year)')
plot_price_trends(pepsico_data, last_year_date, 'PepsiCo - Price Trend (Last Year)')
plot_price_trends(sony_data, last_year_date, 'Sony - Price Trend (Last Year)')
```

```python
# Function to plot price trends using box plots
def plot_price_boxplot(df, start_date, title):
    df['Month'] = df['Date'].dt.to_period('M')
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=df[df['Date'] >= start_date]['Month'].astype(str), y=df[df['Date'] >= start_date]['Price'])
    plt.title(title)
    plt.xlabel('Month')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.show()

# Plot price trends for the last year using box plots
plot_price_boxplot(nike_data, last_year_date, 'Nike - Price Distribution (Last Year)')
plot_price_boxplot(adidas_data, last_year_date, 'Adidas - Price Distribution (Last Year)')
plot_price_boxplot(pepsico_data, last_year_date, 'PepsiCo - Price Distribution (Last Year)')
plot_price_boxplot(sony_data, last_year_date, 'Sony - Price Distribution (Last Year)')
```

```python
# Combine data for correlation analysis
combined_data = pd.concat([nike_data['Price'], adidas_data['Price'], pepsico_data['Price'], sony_data['Price']], axis=1)
combined_data.columns = ['Nike', 'Adidas', 'PepsiCo', 'Sony']

# Calculate correlations
correlation_matrix = combined_data.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Stock Prices')
plt.show()
```

```python
# Ensure the 'Vol.' column is correctly formatted as a numeric value
nike_data['Vol.'] = nike_data['Vol.'].str.replace('M', 'e6').str.replace('K', 'e3')
nike_data['Vol.'] = pd.to_numeric(nike_data['Vol.'], errors='coerce')

adidas_data['Vol.'] = adidas_data['Vol.'].str.replace('M', 'e6').str.replace('K', 'e3')
adidas_data['Vol.'] = pd.to_numeric(adidas_data['Vol.'], errors='coerce')

pepsico_data['Vol.'] = pepsico_data['Vol.'].str.replace('M', 'e6').str.replace('K', 'e3')
pepsico_data['Vol.'] = pd.to_numeric(pepsico_data['Vol.'], errors='coerce')

sony_data['Vol.'] = sony_data['Vol.'].str.replace('M', 'e6').str.replace('K', 'e3')
sony_data['Vol.'] = pd.to_numeric(sony_data['Vol.'], errors='coerce')

# Function to plot price vs. volume
def plot_price_vs_volume(df, start_date, title):
    plt.figure(figsize=(12, 6))
    plt.scatter(df[df['Date'] >= start_date]['Price'], df[df['Date'] >= start_date]['Vol.'], alpha=0.5)
    plt.title(title)
    plt.xlabel('Price')
    plt.ylabel('Volume')
    plt.show()

# Plot price vs. volume for the last year
plot_price_vs_volume(nike_data, last_year_date, 'Nike - Price vs. Volume (Last Year)')
plot_price_vs_volume(adidas_data, last_year_date, 'Adidas - Price vs. Volume (Last Year)')
plot_price_vs_volume(pepsico_data, last_year_date, 'PepsiCo - Price vs. Volume (Last Year)')
plot_price_vs_volume(sony_data, last_year_date, 'Sony - Price vs. Volume (Last Year)')
```
