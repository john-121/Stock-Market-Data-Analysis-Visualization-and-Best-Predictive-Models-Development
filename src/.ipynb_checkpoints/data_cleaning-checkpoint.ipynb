```python
#import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Load uncleaned data
nike_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/nike_data.csv')
adidas_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/adidas_data.csv')
pepsico_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/pepsi_data.csv')
sony_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/sony_data.csv')
```

```python
# Verify the 'Date' column is present in uncleaned data
print("Nike Data Columns:", nike_data.columns)
print("Adidas Data Columns:", adidas_data.columns)
print("PepsiCo Data Columns:", pepsico_data.columns)
print("Sony Data Columns:", sony_data.columns)
```

```python
#parsing dates with each data files
nike_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/nike_data.csv', parse_dates=['Date'], dayfirst=True)
adidas_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/adidas_data.csv', parse_dates=['Date'], dayfirst=True)
pepsico_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/pepsi_data.csv', parse_dates=['Date'], dayfirst=True)
sony_data = pd.read_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/sony_data.csv', parse_dates=['Date'], dayfirst=True)
```

```python
# Strip any leading/trailing whitespace from column names
nike_data.columns = nike_data.columns.str.strip()
adidas_data.columns = adidas_data.columns.str.strip()
pepsico_data.columns = pepsico_data.columns.str.strip()
sony_data.columns = sony_data.columns.str.strip()

# Ensure the 'Date' column is correctly parsed
nike_data['Date'] = pd.to_datetime(nike_data['Date'])
adidas_data['Date'] = pd.to_datetime(adidas_data['Date'])
pepsico_data['Date'] = pd.to_datetime(pepsico_data['Date'])
sony_data['Date'] = pd.to_datetime(sony_data['Date'])
```

```python
# Perform other data cleaning steps
# Dropping duplicates and handling missing values
nike_data.drop_duplicates(inplace=True)
adidas_data.drop_duplicates(inplace=True)
pepsico_data.drop_duplicates(inplace=True)
sony_data.drop_duplicates(inplace=True)

# Verify 'Date' column is still present after cleaning
print('Nike Data Columns after cleaning:', nike_data.columns)
print('Adidas Data Columns after cleaning:', adidas_data.columns)
print('PepsiCo Data Columns after cleaning:', pepsico_data.columns)
print('Sony Data Columns after cleaning:', sony_data.columns)
```

```python
# Handle missing values using forward fill, backward fill, and interpolation 
nike_data.ffill(inplace=True) 
nike_data.bfill(inplace=True) 
nike_data.interpolate(inplace=True) 

adidas_data.ffill(inplace=True) 
adidas_data.bfill(inplace=True) 
adidas_data.interpolate(inplace=True) 

pepsico_data.ffill(inplace=True) 
pepsico_data.bfill(inplace=True) 
pepsico_data.interpolate(inplace=True) 

sony_data.ffill(inplace=True) 
sony_data.bfill(inplace=True) 
sony_data.interpolate(inplace=True) 


# Print the first few rows to verify 
print(nike_data.head()) 
print(adidas_data.head()) 
print(pepsico_data.head())
print(sony_data.head())
```

```python
#Address Outliers
from scipy.stats.mstats import winsorize

# Function to apply Winsorization
def winsorize_column(df, column):
    df[column] = winsorize(df[column], limits=[0.01, 0.01])

# Apply Winsorization to relevant columns
for column in ['Price', 'Open', 'High', 'Low']:
    winsorize_column(nike_data, column)
    winsorize_column(adidas_data, column)
    winsorize_column(pepsico_data, column)
    winsorize_column(sony_data, column)

# Print the first few rows to verify
print(nike_data.head())
print(adidas_data.head())
print(pepsico_data.head())
print(sony_data.head())
```

```python
# Clean column names by striping leading/trailing whitespace
nike_data.columns = nike_data.columns.str.strip()
adidas_data.columns = adidas_data.columns.str.strip()
pepsico_data.columns = pepsico_data.columns.str.strip()
sony_data.columns = sony_data.columns.str.strip()
```

```python
# Ensure the 'Date' column is correctly formatted
nike_data['Date'] = pd.to_datetime(nike_data['Date'])
adidas_data['Date'] = pd.to_datetime(adidas_data['Date'])
pepsico_data['Date'] = pd.to_datetime(pepsico_data['Date'])
sony_data['Date'] = pd.to_datetime(sony_data['Date'])

# Print the first few rows to verify
print(nike_data.head())
print(adidas_data.head())
print(pepsico_data.head())
print(sony_data.head())
```

```python
# Check for missing values and duplicates
print(nike_data.isnull().sum())
print(adidas_data.isnull().sum())
print(pepsico_data.isnull().sum())
print(sony_data.isnull().sum())

# Remove duplicates
nike_data.drop_duplicates(inplace=True)
adidas_data.drop_duplicates(inplace=True)
pepsico_data.drop_duplicates(inplace=True)
sony_data.drop_duplicates(inplace=True)
```

```python
# Save cleaned data 
nike_data.to_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_nike_data.csv', index=False)
adidas_data.to_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_adidas_data.csv', index=False)
pepsico_data.to_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_pepsi_data.csv', index=False)
sony_data.to_csv('C:/Users/johnt/Stock-Market-Data-Analysis-Visualization-and-Best-Predictive-Models-Development/datasets/cleaned_sony_data.csv', index=False)
```
